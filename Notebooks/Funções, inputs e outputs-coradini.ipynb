```python
from pandas_datareader import data as web

TickerA='ITUB3.SA'
TickerB='BBDC3.SA'
TickerC='BBAS3.SA'
TickerD='SANB11.SA'
TickerE='BOVA11.SA'

df=pd.DataFrame()
tickers = [TickerA, TickerB, TickerC, TickerD, TickerE]
for t in tickers:
    df[t] = web.get_data_yahoo(t,'01/01/2010')['Adj Close']
```

```python
df.head()
```

```python
df.rename(columns ={'ITUB3.SA':'ITUB', 'BBDC3.SA':'BBDC','BBAS3.SA':'BBAS','SANB11.SA':'SANB', 'BOVA11.SA':'BOVA'},inplace = True)
df.reset_index(inplace = True)
df.dropna(inplace=True)
```

```python
from matplotlib import pyplot as plt

tickers = list(df.drop(['Date','BOVA'], axis = 1).columns)
plt.figure(figsize=(16,6))

for i in tickers:
        plt.plot(df['Date'], df[i])
plt.legend(tickers)
plt.title("Cotação x tempo", fontsize = 25)
plt.grid()
plt.show()
```
