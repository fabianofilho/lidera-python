### 1. Pergunte ao usuário qual dia da semana é hoje. Utilize condicional if/else para dizer se este dia é fim de semana ou não.

```python
dias_de_semana = ['segunda', 'terça', 'quarta', 'quinta', 'sexta']

dia = input('Que dia é hoje? ')
dia = dia.lower()

if dia in dias_de_semana:
    print('Dia da semana')
else:
    print('Fim de semana :)')
```

### 4. Crie uma função que calcula e printa o fatorial de um número
4 fatorial (4!) = 1 x 2 x 3 x 4 = 24


```python
def fatorial(numero):
    fat = 1
    for i in range(numero):
        i += 1
        fat = fat * i
    return fat
```

```python
fatorial(4)
```

### 7. Crie uma função que recebe dois números como parâmetros e retorna o maior deles

```python
lista7 = [10,500]

def maximo(seq):
  valor_maximo = seq[0]
  for i in lista7:
    if i > valor_maximo:
      valor_maximo = i
  return valor_maximo
```

### 14. Crie uma função que confere se um número é primo

```python
def primo(num):

  metade = int(num/2)

  for i in range(metade):
    i = i+2
    if num % i == 0:
      return 'Falso'
  return 'Verdadeiro'
```

```python
primo(15)
```

### 15. Utilize a função acima para criar uma função que gera uma lista de números primos de 0 até um número fornecido (parâmetro da função)

```python
def lista_primos(maximo):
  lista = []
  for i in range(maximo):
    i += 1
    if primo(i) == 'Verdadeiro':
      lista.append(i)
  return lista
```

```python
lista_primos(100)
```
