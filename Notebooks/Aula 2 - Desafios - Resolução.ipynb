**Desafio 2**

Dada as listas:

lista1 = [20, 34, 23, 2, 45]

lista2 = [33, 67, 83, 29, 40]

Coloque a lista1 em ordem decrescente e a lista2 em ordem crescente.

De forma manual faça uma lista de pares chave-valor de tuplas, onde:

Chave = O algarismo do binário do item da lista1

Valor = A soma dos algarismos dos binários dos itens da lista 1 e da lista2 relativos a mesma posição

A partir disso crie um dicionário e retorne o valor da chave = "10"

```
lista1 = [20, 34, 23, 2, 45]
lista2 = [33, 67, 83, 29, 40]

lista1.sort(reverse=True)
lista2.sort()

chave0 = int(str(bin(lista1[0]))[2:])
chave1 = int(str(bin(lista1[1]))[2:])
chave2 = int(str(bin(lista1[2]))[2:])
chave3 = int(str(bin(lista1[3]))[2:])
chave4 = int(str(bin(lista1[4]))[2:])

valor0 = int(str(bin(lista2[0]))[2:]) + chave0
valor1 = int(str(bin(lista2[1]))[2:]) + chave1
valor2 = int(str(bin(lista2[2]))[2:]) + chave2
valor3 = int(str(bin(lista2[3]))[2:]) + chave3
valor4 = int(str(bin(lista2[4]))[2:]) + chave4

dicionario = dict([(chave0, valor0), (chave1, valor1), (chave2, valor2), (chave3, valor3), (chave4, valor4)])
print("O valor da chave igual a '10' é " + str(dicionario[10]))
```

**Desafio 3**

Ordene o texto, retornem os indexs das primeiras ocorrências das letra "a", "e", "i", "o", "u" e "!" na variável "texto" em uma única variável (variável x). Após isso faça as seguintes operações:

(1) - Faça uma adição da variável por ela mesma

(2) - Faça uma exponenciação da variável por 2

(3) - Encontre o módulo do valor dela pelo número ao contrário dela mesma

Depois faça:

(2) - (1) + (3)



```
texto = "Olá Lidera. Eu estou feliz em tê-los conosco!"
lista = list(texto)
lista.sort()
a = lista.index("a")
e = lista.index("e")
i = lista.index("i")
o = lista.index("o")
u = lista.index("u")
exc = lista.index("!")

palavra = int(str(a) + str(e) + str(i) + str(o) + str(u) + str(exc))

adicao = palavra + palavra
exponencial = palavra ** 2

arvalap = list(str(palavra))
arvalap.reverse()
arvalap1 = int(arvalap[0] + arvalap[1] + arvalap[2] + arvalap[3] + arvalap[4] + arvalap[5] + arvalap[6] + arvalap[7] + arvalap[8] + arvalap[9] + arvalap[10])
modulo = palavra % arvalap1

print(
    "Adição = " + str(adicao)
    + "\nExponencial = " + str(exponencial)
    + "\nModulo = " + str(modulo)
)

resultado = exponencial - adicao + modulo
print("Resultado = " + str(resultado))
```
