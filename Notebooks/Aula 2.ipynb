```
# Algumas formas de utilizar listas:
# lista = [0] * 4
# print(lista)

# lista = lista + [1, 2] * 3
# print(lista)

lista = ["ovo", "leite", 3, 2, True, "ola"]
print(lista[1:]) # Posição 1 até o final
print(lista[:4]) # Início até antes da posição 4
print(lista[1:4]) # Posição 1 até antes da posição 4
print(lista[1:-1]) # Posição 1 até a penúltima posição
print(lista[-1])

del lista[2] # Vai excluir item na posição 2
print(lista)
```

```
# Atribuição a fatias

lista = [1, 'y', ['ab', 'CD']]
print(lista[1:1])
print(lista[1])
# lista[1:1] é diferente de lista[1]
# lista[1:1] => Começa da posição 1 e vai até a posição 1, sem incluí-la
lista[1:1] = ['z']
print(lista)
# Demonstrar ao vivo

lista[1:3] = [['x']]
print(lista)
# Demonstrar ao vivo

lista[1:-1] = [2, 3, 4]
print(lista)

lista[:-2] = 'xyz'
print(lista)
# Se a intenção é colocar um vetor/string por inteiro em uma posição
# é necessário que seja incrementado como lista
# Demonstrar ao vivo
```

```
# Incremento em Fatias

lista = ['a', 2, 3, 'd', 'x']
print(lista[::2])
# Ele pega de 2 em 2 da lista inteira
# Demonstrar ao vivo

print(lista[:3:2])
# Ele faz primeiro uma lista[:3]
# Dessa lista[:3] ele pega de 2 em 2
# Demonstrar ao vivo

print(lista[::-1])
# Ele pega a lista e mostra ela de -1 em -1
# Demonstrar ao vivo
```

```
# Principais métodos das listas

# list
listanova = list([1, 2, "ovo", 1.5, True, False, "leite", [1, 2, 4, 7, 7, 9]])
print(listanova)


lista = []

# append
lista = [1, 2, 3, 4]
lista.append(5) # Adiciona um item ao final
print(lista)

# clear
lista.clear() # Apaga a lista
print(lista)

# copy
lista = [1, 2, 3, 4]
lista2 = lista
print("Lista é = " + str(lista))
print("Lista2 é = " + str(lista2))
lista.clear()
print("Depois do .copy Lista é = " + str(lista))
print("Depois do .copy Lista2 é = " + str(lista2))
# A lista2 é alterada em conjunto com as alterações da lista
# isso se dá porque o que foi passado para a lista2 não foi as informações da lista
# e sim o "caminho" da lista

lista = [1, 2, 3, 4]
lista2 = lista.copy()
print("Lista é = " + str(lista))
print("Lista2 é = " + str(lista2))
lista.clear()
print("Depois do .copy Lista é = " + str(lista))
print("Depois do .copy Lista2 é = " + str(lista2))

# count
lista = [1, 2, 3, 2, 4, 2, 5]
print("A quantidade de 2 na lista é " + str(lista.count(2)))

# extend
lista = [1, 2, 3, 4]
lista.append([5, 6])
print("Lista com .append " + str(lista))

lista = [1, 2, 3, 4]
lista.extend([5, 6])
print(lista)
print("Lista com .extend " + str(lista))
# Adiciona-se uma lista como elementos iteráveis

# index
lista = ["ovo", "leite", "ovo", "queijo"]
print(lista.index("queijo")) # retorna a primeira posição da informação

# insert
lista = ["ovo", "leite", "ovo", "queijo"]
lista.insert(3, "abacaxi") # (número da posição a ser inserido, informação)
# Adicionou "abacaxi" na posição 3 da lista
print(lista)

# pop
lista = ["ovo", "leite", "ovo", "queijo"]
lista.pop(2) # Remove a informação da posição 2
print(lista)

# remove
lista = ["ovo", "leite", "ovo", "queijo"]
lista.remove("ovo") # Remove a primeira ocorrência de "ovo" na lista
print(lista)

# reverse
lista = ["ovo", "leite", "ovo", "queijo"]
lista.reverse() # Vai reverter a ordem da lista
print(lista)

# sort
lista = ["ovo", "leite", "ovo", "queijo"]
lista.sort() # Vai organizar a lista de forma crescente
print(lista)

lista = ["ovo", "leite", "ovo", "queijo"]
lista.sort(reverse=True) # Vai organizar a lista de forma decrescente
print(lista)
```

```
# Principais métodos

# fromkeys
x = ("chave1", "chave2", "chave3")
y = 0
dicionario = dict.fromkeys(x,y) # Atribui valor de y para todas as chaves
# É possível declarar sem y
print(dicionario)

# get
dicionario = {
      "João": 27911111111, 
      "Maria": 27922222222,
      "Cecília": 27933333333
}
print(dicionario.get("Maria"))
# Mostra o valor da chave especificada

# items
dicionario = {
      "João": 27911111111, 
      "Maria": 27922222222,
      "Cecília": 27933333333
}
print(dicionario.items())
# lista todas as duplas chave-valor em tuplas

# keys
dicionario = {
      "João": 27911111111, 
      "Maria": 27922222222,
      "Cecília": 27933333333
}
print(dicionario.keys())
# lista todas as chaves do dicionário

# pop
dicionario = {
      "João": 27911111111, 
      "Maria": 27922222222,
      "Cecília": 27933333333
}
dicionario.pop("João") # Retira a informação chave-valor referente a chave
print(dicionario)

# popitem
dicionario = {
      "João": 27911111111, 
      "Maria": 27922222222,
      "Cecília": 27933333333
}
dicionario.popitem() # Remove o último par chave-valor
print(dicionario)

# setdefault
dicionario = {
      "João": 27911111111, 
      "Maria": 27922222222,
      "Cecília": 27933333333
}
print(dicionario.setdefault("João")) # Retorna o valor da chave

dicionario = {
      "João": 27911111111, 
      "Maria": 27922222222,
      "Cecília": 27933333333
}
print(dicionario.setdefault("Agripina")) # Se a chave não existir, insere a informação chave-valor
print(dicionario)

dicionario = {
      "João": 27911111111, 
      "Maria": 27922222222,
      "Cecília": 27933333333
}
print(dicionario.setdefault("Agripina", 27944444444))
print(dicionario)

# update
dicionario = {
      "João": 27911111111, 
      "Maria": 27922222222,
      "Cecília": 27933333333
}
dicionario.update({"Agripina": 27944444444}) # Adiciona ao dicionario um par chave-valor
print(dicionario)

# values
dicionario = {
      "João": 27911111111, 
      "Maria": 27922222222,
      "Cecília": 27933333333
}
print(dicionario.values()) # Lista todos os valores do dicionario
```
