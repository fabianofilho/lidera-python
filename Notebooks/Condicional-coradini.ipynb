# ELIF:

```python
# Maior que 7 = aprovado
# Menor que 5 = reprovado
# Entre 5 e 7 = prova final

nota = 8

if nota >= 7:
    print ("Aprovado")
else: 
    if nota < 5:
        print ("Reprovado")
    else:
        print ("Prova Final")
```

```python
nota = 3

if nota >= 7:
    print ("Aprovado")
elif nota < 5:
    print ("Reprovado")
elif 5 < nota < 7:
    print ("Prova Final")
```

```python
lista_de_compras = ["Maçã",'Abacaxi','Pera']

pertence = []

if "Maçã" in lista_de_compras:
    pertence.append("Maçã")
elif "Abacaxi" in lista_de_compras:
    pertence.append("Abacaxi")

print(pertence)
```

```python
lista_de_compras = ["Maçã",'Abacaxi','Pera']
pertence = []

if "Maçã" in lista_de_compras:
    pertence.append("Maçã")
if "Abacaxi" in lista_de_compras:
    pertence.append("Abacaxi")

print(pertence)
```

```python
nota = int(input("Digite sua nota: "))

if nota >= 7:
    print ("Aprovado")
elif nota < 5:
    print ("Reprovado")
else:
    print("Prova Final")
```

```python
# or

idade = 10

if idade <= 5 or idade >= 65:
    print ("Não paga passagem de ônibus")
else:
    print("Paga passagem de ônibus")
```

```python
# if not

idade = 10

if not (5 < idade < 65):
    print ("Não paga passagem de ônibus")
else:
    print("Paga passagem de ônibus")
```

```python
fruta = "Maçã"
cor = "Vermelha"

if fruta == "Maçã" and cor == "Verde":
    print ("É uma maçã verde!")
    
elif fruta == "Maçã" and cor != "Verde":
    print("É uma maçã, mas não é verde")
    
else: print("Não é uma maçã!")
```

```python
cores_arco_iris = ['Vermelho','Laranja','Amarelo','Verde','Azul','Anil','Violeta']

if 'Azul' in cores_arco_iris:
    print("Azul é uma cor do arco-íris!")
```
