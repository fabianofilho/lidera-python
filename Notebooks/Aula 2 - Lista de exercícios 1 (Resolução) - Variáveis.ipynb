**Exercício 1**

x = 10

y = 5

O objetivo é passar o valor inicial de y para x e o valor inicial de x para y, usando **SOMENTE** manipulação de variáveis e não de valores. Ao final deve-se imprimir os valores.

Pode-se criar quantas variáveis necessárias, desde que não se manipule valores nelas.

Ou seja, no final, esperamos que:

x = 5

y = 10

------------------------------------

**É PROIBIDO:**

Mudar manualmente os valores de x e y.

Não é permitida a simples atribuição de x = 5 e y = 10

```
x = 10
y = 5
z = x # Nova variável criada só para receber "x"
x = y
y = z

print("x = " + str(x))
print("y = " + str(y))
```

**Exercício 2**

Usando apenas **2 variáveis tipo int e 1 variável tipo string** deve-se imprimir uma mensagem com os seguintes dizeres:

"Olá Lidera! Meu nome é *SEU_NOME*, tenho *SUA_IDADE* anos e minha matrícula é *SUA_MATRICULA*"

Em seguida deve-se imprimir em uma mesma função de impressão os valores de *SEU_NOME*, *SUA_IDADE* e *SUA_MATRICULA* cada um em uma linha. Espera-se então que a mensagem seja:

*SEU_NOME*

*SUA_IDADE*

*SUA_MATRICULA*

```
seuNome = "Lucas Ferrari Corrêa"
suaIdade = 24
suaMatricula = 2014100692

print(
    "Olá Lidera! Meu nome é "
    + seuNome
    + ", tenho "
    + str(suaIdade)
    + " anos e minha matrícula é "
    + str(suaMatricula)
)

print(
    seuNome
    + "\n"
    + str(suaIdade)
    + "\n"
    + str(suaMatricula)
)
```

**Exercício 3**

Retornem os indexs das primeiras ocorrências das letra "a", "e", "i", "o", "u" e "!" na variável "texto" e em uma única opção de imprimir, imprimam-nos juntos.

Espera-se então:

975181344



```
texto = "Olá Lidera. Eu estou feliz em tê-los conosco!"

print(
    str(texto.index("a"))
    + str(texto.index("e"))
    + str(texto.index("i"))
    + str(texto.index("o"))
    + str(texto.index("u"))
    + str(texto.index("!"))
)
```
